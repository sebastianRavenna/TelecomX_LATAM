```python
# Extraer los diccionarios en columnas nuevas
customer_flat = pd.json_normalize(df_Telecom['customer'])
phone_flat = pd.json_normalize(df_Telecom['phone'])
internet_flat = pd.json_normalize(df_Telecom['internet'])
account_flat = pd.json_normalize(df_Telecom['account'])
```

```python
df_Telecom = pd.concat(
    [df_Telecom[['customerID', 'Churn']], customer_flat, phone_flat, internet_flat, account_flat],
    axis=1
)
df_Telecom.head()
```

```python
import numpy as np
```

```python
df_Telecom['Charges.Total'] = pd.to_numeric(df_Telecom['Charges.Total'], errors='coerce')
df_Telecom['Charges.Total'].isnull().sum()
```
